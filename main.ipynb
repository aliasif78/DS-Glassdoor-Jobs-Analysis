# Verifying if Title matches Description

```python
df['Title'].value_counts()
```

```python
verification = []

for titleEntry, descriptionEntry in zip(df['Title'], df['Description']):
    if (('data scientist' in titleEntry.lower() and 'data scientist' in descriptionEntry.lower()) or
        ('data analysis' in titleEntry.lower() and 'data analysis' in descriptionEntry.lower()) or
        ('data science' in titleEntry.lower() and 'data science' in descriptionEntry.lower()) or
        ('data analyst' in titleEntry.lower() and 'data analyst' in descriptionEntry.lower()) or
        ('data analytics' in titleEntry.lower() and 'data analytics' in descriptionEntry.lower()) or
        ('data engineer' in titleEntry.lower() and 'data engineer' in descriptionEntry.lower())):
        verification.append('No')
    else:
        verification.append('Yes')
```

```python
verification
```

```python
df['Verification Warning'] = verification
df
```
